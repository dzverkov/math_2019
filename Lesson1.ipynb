### Задание 3

Напишите код на Python, реализующий построение графиков:
1.	окружности,
2.	эллипса,
3.	гиперболы.

```python
x = []
y1 = []
y2 = []
R = 3

for i in range(1000*2*R+1):
    x1 = -R + i/1000
    x.append(x1)
    y1.append(math.sqrt(R**2 - x1**2))
    y2.append(-math.sqrt(R**2 - x1**2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-4, 4)
plt.xlim(-5.7, 5.7)
plt.grid(True)
plt.show()
```

```python
x = []
y1 = []
y2 = []
a = 5
b = 3

for i in range(1000*2*a+1):
    x1 = -a + i/1000
    x.append(x1)
    y1.append(math.sqrt(1 - x1**2/a**2) * b)
    y2.append(-math.sqrt(1 - x1**2/a**2) * b)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-4, 4)
plt.xlim(-5.7, 5.7)
plt.grid(True)
plt.show()
```

```python
x = []
y1 = []
y2 = []
a = 1
b = 3

for i in range(1000*2*a):
    x1 = a + i/1000
    x.append(x1)
    y1.append(math.sqrt(x1**2/a**2 - 1) * b)
    y2.append(-math.sqrt(x1**2/a**2 - 1) * b)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-4, 4)
plt.xlim(-5.7, 5.7)
plt.grid(True)
plt.show()
```

### Задание 4

Нарисуйте трехмерный график двух параллельных плоскостей

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 3*Y
Z2 = 2*X + 3*Y + 10
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
plt.show()
```

Нарисуйте трехмерный график двух любых поверхностей второго порядка

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-20, 20, 1)
Y = np.arange(-20, 20, 1)
X, Y = np.meshgrid(X, Y)
Z = X**3 + Y**3
Z2 = X**2 + Y**2
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
plt.show()
```
