### Задание 1

Известно, что генеральная совокупность распределена нормально  
со средним квадратическим отклонением, равным 16.  
Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,  
если выборочная средняя M = 80, а объем выборки n = 256.

```python
std_ = 16 # среднее квадратичное отклонение
Mx = 80 # выборочное среднее
n = 256 # объем выборки
a = 5 # уровень значимости (100%-95%)
Za = 1.96

p1 = Mx - Za *std_/np.sqrt(n)
p2 = Mx + Za *std_/np.sqrt(n)

print(p1,p2)
```

### Задание 2

В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,  
получены опытные данные:  
6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1  
Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,  
оценить истинное значение величины X при помощи доверительного интервала, покрывающего это  
значение с доверительной вероятностью 0,95.

```python
ar = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
std_ = np.std(ar, ddof=1) # среднее квадратичное отклонение
Mx = ar.mean() # выборочное среднее
n = len(ar) # объем выборки
a = 5 # уровень значимости (100%-95%)
Ta = 2.262 # Для n-1=9 и a=0.025

p1 = Mx - Ta *std_/np.sqrt(n)
p2 = Mx + Ta *std_/np.sqrt(n)

print(p1,p2)
```

### Задание 3

Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний диаметр 17 мм.
Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр
оказался равным 17.5 мм, а дисперсия известна и равна 4 мм.

```python
mu0 = 17    # среднее значение для гипотезы 0
mu1 = 17.5  # среднее значение для гипотезы 1
a = 0.05   # односторонний критерий
n = 100    # объем выборки
D = 4      # дисперсия

# H0 : mu0 = mu - предположение по гипотезе 0
# H1 : mu1 > mu0 - предположение по гипотезе 1

Zt = 1.645 # Z табличное для 0.995
Zr = (mu1 - mu0) / (np.sqrt(D)/np.sqrt(n)) # Z расчётное 

print(f"Zt = {Zt}")
print(f"Zr = {Zr}")

if Zr > Zt:
    print("Гипотеза H0 не подтверждена, H1 верна")
else:
    print("Гипотеза H1 подтверждена, H1 не верна")
```

### Задание 4

Продавец утверждает, что средний вес пачки печенья составляет 200 г.  
Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет:  
202, 203, 199, 197, 195, 201, 200, 204, 194, 190.  
Известно, что их веса распределены нормально.  
Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%?  

```python
ar = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
std_ = np.std(ar, ddof=1) # среднее квадратичное отклонение

mu0 = 200 # среднее значение для гипотезы 0
mu1 = ar.mean()  # среднее значение для гипотезы 1
a = 0.001   # двухсторонний критерий (100% - 99%)
n = len(ar)    # объем выборки
D = np.var(ar, ddof=1) # дисперсия

# H0 : mu0 = mu - предположение по гипотезе 0
# H1 : mu1 != mu0 - предположение по гипотезе 1

Tt = 3.250 # Табличное T, для n-1=9 и a=0.005
Tr = (mu1 - mu0) / (np.sqrt(D)/np.sqrt(n)) # T расчётное 

print(f"Tt = {Tt}")
print(f"Tr = {Tr}")

if -Tt <= Tr and Tr <= Tt:
    print("Гипотеза H0 подтверждена, продавец прав")
else:
    print("Гипотеза H0 не подтверждена, продавец не прав")
```
